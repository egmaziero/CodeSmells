```python
import pandas as pd
```

```python
dataset = pd.read_csv('../data/dataset_badsmells.csv', delimiter=';', index_col=0)
print(dataset.shape)
```

```python
def replace(text):
    path1 = '/home/matheus/Documentos/repositories/'
    path2 = '/home/matheus/Documentos/repositories/'
    text = text.replace(path1,'')
    text = text.replace(path2,'')
    return text

def find_lang(text):
    return text.split('/')[0]

def find_app(text):
    return text.split('/')[0]

dataset['file'] = dataset['file'].apply(replace)
dataset['lang'] = dataset['file'].apply(find_lang)
dataset = dataset[dataset['lang']=='Java']  # keep only apps in Java
```
